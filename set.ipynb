# Sets

Sets are mutable container object with no duplicate values and indexing support.

## Empty Set Object

```python
set()  # Set constructor
```

## Non-empty Set Object

```python
x = {1, 2, 'a', 'b'}
```

```python
type(x)
```

### No duplicates allowed

```python
y = {1, 2, 3, 4, 1, 3, 8}
print(y)
```

### No Mutable Objects are allowed

```python
z = {1, 2, [3, 4], {5, 6}, {'a': 1, 'b': 2}}
```

```python
y
```

### No Indexing Support

```python
y[-1]
```

## Demonstration of membership testing in list and set

```python
list_object = list(range(10_00_000))
```

```python
list_object[-5:]
```

```python
set_object = set(range(10_00_000))
```

```python
%%timeit
10 in list_object
```

```python
%%timeit
10 in set_object
```

```python
%%timeit
999999 in list_object
```

```python
%%timeit
999999 in set_object
```

```python
dict_object = {i: i for i in range(10_00_000)}
```

```python
%%timeit
999999 in dict_object
```
