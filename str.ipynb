## Basic Properties

* **Mutability**: `str` object can't be modified in place. So, they are immutable objects.
* **Ordered**: Yes (characters in a `str` object have a specific order)
* **Allow duplicates**: Yes (same character can appear multiple times)
* **Indexing support**: Yes (a specific character can be accessed using index)

## Accessing String Objects

### Indexing

```python
language = "Python"
```

```python
# In Python, we can use positive or negative indexing to access elements of str objects

# Positive indexing (starts from 0)
print(language[0])  # P
print(language[1])  # y
print(language[5])  # n
```

```python
# Negative indexing (starts from -1 at the end of the str object)
print(language[-1])  # n
print(language[-2])  # o
print(language[-6])  # P
```

### Slicing

```python
session = "Python Programming"
```

```python
session[0:8:-1]  # session[0:10:1]
```

```python
session[8]
```

```python
# start_index = 0, end_index = 17, step_index = -1
# When step index is negative, start_index = 17, end_index = 0, step_index = -1
session[2:12:-1]  # "gnimmargorP n"
```

```python
session[17::-1]
```

```python
session[11:3:1]
```

```python
session[:25]
```

```python
# Basic slicing [start:end:step]
# start index is accessed but end index is not accessed. 
# step defaults to 1 which means move 1 index at a time.
print(session[0:6]) # Python
```

```python
# When start index is unspecified, Python starts indexing from beginning
print(session[:6])
```

```python
# When end index is unspecified, Python goes until the last character.
# So, character at the last index is also accessed.
print(session[0:])
```

```python
session = "Python Programming"
```

```python
session[-11]
```

```python
# Python also supports negative indexing.
# The last index is -1, second to last is -2 and so on.
# Remember that as end index is not specified, Python goes until the last character.
print(session[-11:2:-1]) 
```

```python
# As start index is not specified, Python starts from the beginning.
print(session[:-10])
```
