# Tuples

Tuples are also container object but it's immutables meaning that once the object is created we cannot modify
it as the string objects.

## Empty tuple object

```python
()  # Using parantheses
```

```python
type(())
```

```python
tuple()  # Using constructor
```

## Non-empty tuple object

```python
(1, 3.5, "data", [3, 4, 5], (1, 2))
```

```python
(1)
```

```python
x = ("a")
type(x)
```

```python
# single element tuple object
y = (1,)
type(y)
```

```python
str_object = ("data analyst",)
type(str_object)
```

## Indexing

```python
objects = (1, 3.5, "data", [3, 4, 5], (1, 2))
```

```python
objects[-1]
```

```python
objects[3][1]  # Multi-indexing
```

## Slicing

```python
objects
```

```python
objects[3:8]
```

```python
objects[3:8:-1]
```

```python
objects[5::-1]  # As end index isn't specified, python will move until it gets the first element.
```

```python
objects[6:5:-1]
```

```python
objects[-5:-3:-1]
```

## Common methods

```python
# index
objects.index((1, 2))
```

## Operators

```python
# `+` operator
(1, 2, 3) + (4, 5, 6)
```

```python
x = [1, 2, 3]
y = [4, 5, 6]

x + y
```

```python
x
```

```python
y
```

```python
# `*` operator
(2, 3, 5) * 3
```

```python
(1, 2, (3, 5)) * 2
```

```python
((1, 2) * 2) * 2
```

## Membership Operators

```python
2 in (1, 2, 3)
```

```python
3 not in (1, 2)
```
